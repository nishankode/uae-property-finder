```python
import pandas as pd
pd.pandas.set_option('display.max_columns', None)
import numpy as np

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('../data/cleaned_data/uae_realestate_old_cleaned.csv')
```

```python
df
```

```python
# Renaming column names to snake case
df = df.rename(columns={
    "Regis": "reg_date",
    "ProcedureValue": "procedure_value",
    "Master Project": "master_project",
    "Master Project Land": "master_project_land",
    "Project": "project",
    "Project Lnd": "project_land",
    "Building No": "building_no",
    "BuildingNameEn": "building_name_en",
    "Size": "size_sqmt",
    "UnitNumber": "unit_number",
    "DmSubNo": "dm_sub_no",
    "PropertyTypeEn": "property_type_en",
    "LandNumber": "land_number",
    "ProcedurePartyTypeNameEn": "procedure_type",
    "NameEn": "name_en",
    "Mobile": "mobile",
    "ProcedureNameEn": "procedure_name",
    "CountryNameEn": "country",
    "IdNumber": "id_number",
    "UaeIdNumber": "uae_id_number",
    "BirthDate": "birth_date"
})
```

```python
# Cleaning Mobile Numbers
all_mobiles = df.groupby('name_en')['mobile'].agg(list).to_frame('mobile_numbers').reset_index()
all_mobiles['mobile_numbers'] = all_mobiles['mobile_numbers'].apply(lambda x: np.nan if len(list(set(x))) == 1 and pd.isnull(list(set(x))[0]) else list(set(x)))
all_mobiles = all_mobiles[all_mobiles['mobile_numbers'].notna()]

# Cleaning Date
df['reg_date'] = pd.to_datetime(df['reg_date'])

# Creating size_sqft
df['size_sqft'] = df['size_sqmt'].apply(lambda x: round(x * 10.76391, 2))
```

# 1. Name Search

```python
def search_name(name, df):
    
    matches = df[df['name_en'] == name]
    
    return matches
```

```python
name_filter_result = search_name('DAOUD YAQUB', df)
name_filter_result
```

# 2. ID Search

```python
def search_id(_id, df):

	matches = df[(df['id_number'] == _id) | (df['uae_id_number'] == _id)]

	return matches
```

```python
id_filter_result = search_id('530946649', df)
id_filter_result
```

# 3. Phone Number Search

```python
def search_phone(mobile, df):
    
    matches = df[df['mobile'] == mobile]

    return matches
```

```python
mobile_search_result = search_phone('971-50-6544384', df)
mobile_search_result
```

# 4. Property Search

```python
def search_property(property_name=None, df=None, size_sqmt=None, size_sqft=None, transaction_date=None, unit_number=None):

	all_matches = df[
		(df['master_project'] == property_name) |
		(df['master_project_land'] == property_name) |
		(df['project'] == property_name) |
		(df['project_land'] == property_name) |
		(df['building_name_en'] == property_name)
		]

	if size_sqmt:
		all_matches = all_matches[all_matches['size_sqmt'] == size_sqmt]

	if size_sqft:
		all_matches = all_matches[all_matches['size_sqft'] == size_sqft]
	
	if transaction_date:
		all_matches = all_matches[all_matches['reg_date'] == transaction_date]

	if unit_number:
		all_matches = all_matches[all_matches['unit_number'].apply(lambda x: x.endswith(unit_number))]

	return all_matches
```

```python
# Without Size and Date
property_search_result = search_property(property_name='OCEAN HEIGHTS', df=df)
property_search_result
```

```python
# With Size and Date
property_search_result = search_property(property_name='OCEAN HEIGHTS', df=df, size_sqft=1598.01, transaction_date='2018-10-02')
property_search_result
```

```python
# Without Size and Date, With Unit Number
property_search_result = search_property(property_name='PRINCESS TOWER', df=df, unit_number='06')
property_search_result
```
